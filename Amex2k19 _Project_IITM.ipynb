###### As thereâ€™s inconsistency in the way missing values are entered in the dataset, which are given using both â€˜naâ€™ and â€˜missingâ€™, they are all replaced with â€˜NaNâ€™ values by using the replace( ) function of the NumPy library.

```python
# lets resolve that issue 
for col in data.columns.values.tolist():
    if data[col].dtypes == 'object':
        data[col].replace(['na','missing'],[np.nan,np.nan], inplace=True)
        data[col]=data[col].astype("float")   

for col in test_data.columns.values.tolist():
    if test_data[col].dtypes=='object':
        test_data[col].replace(['na','missing'],[np.nan,np.nan], inplace=True)
        test_data[col]=test_data[col].astype("float")
```

# Feature Engineering

With our level of domain knowledge in credit card terms we have added four additional variables as columns to the datasets.

- By adding the variables â€˜mvar3â€™, â€˜mvar4â€™, â€˜mvar5â€™ we created a variable called â€˜Severity_sumâ€™
Similarly, we have, 
- â€˜num_active_fullâ€™ = â€˜mvar17â€™ + â€˜mvar18â€™

- â€˜num_active_75%â€™ = â€˜mvar19â€™ + â€˜mvar20â€™

```python
df=data

df['Total_Severity']=df['mvar3']+df['mvar4']+df['mvar5']

df['max_credit_avail']=df['mvar7']+df['mvar8']

df['total_num_avail']= df['mvar17']+df['mvar18']

df['total_75%_credit']= df['mvar19']+df['mvar20']
```

```python
test_data['Total_Severity']=test_data['mvar3']+test_data['mvar4']+test_data['mvar5']

test_data['max_credit_avail']=test_data['mvar7']+test_data['mvar8']

test_data['total_num_avail']= test_data['mvar17']+test_data['mvar18']

test_data['total_75%_credit']= test_data['mvar19']+test_data['mvar20']
```


# Model Training and Evaluation

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold, train_test_split
kfold = StratifiedKFold(n_splits=5)
from sklearn.metrics import roc_auc_score,roc_curve,auc
from imblearn.over_sampling import SMOTE
```

```python
Y_train = df["default_ind"]
X_train = df.drop(["default_ind","application_key"], axis=1)
X_train.shape, Y_train.shape
```

 - The training dataset has been split into train and validation data by assigning 20% into test_size parameter of the train_test_split( ) function.

```python
train_x,val_x,train_y,val_y = train_test_split(X_train, Y_train, test_size = 0.20, random_state=14)
train_x.shape,val_x.shape,train_y.shape,val_y.shape
```


## LightBGM Model

```python
#pip install lightgbm
```

```python
import lightgbm
train_data = lightgbm.Dataset(train_x, label=train_y)
val_data = lightgbm.Dataset(val_x, label=val_y)

para={'boosting_type': 'gbdt',
 'colsample_bytree': 0.65,
 'learning_rate': 0.005,
 'max_bin': 512,
 'max_depth': -1,
 'metric': 'auc',
 'min_child_samples': 5,
 'min_child_weight': 1,
 'min_split_gain': 0.5,
 'nthread': 3,
 'num_class': 1,
 'num_leaves': 23,
 'objective': 'binary',
 'reg_alpha': 0.8,
 'reg_lambda': 1.2,
 'scale_pos_weight': 1,
 'subsample': 0.7,
 'subsample_for_bin': 200,
 'subsample_freq': 1}

```

```python
model = lightgbm.train(para,
                       train_data,
                       valid_sets=val_data,
                       num_boost_round=5000,
                       early_stopping_rounds=50,
                      )
```

```python
app_key1=test_data.application_key
test_data=test_data.drop("application_key",axis=1)
x = test_data.values
y_pred1 = model.predict(x)
a = np.where(y_pred1 > 0.5, 1, 0)
```

```python
print("Importance of features...")
gain = model.feature_importance('gain')
ft = pd.DataFrame({'feature':model.feature_name(), 'split':model.feature_importance('split'), 'gain':100 * gain / gain.sum()}).sort_values('gain', ascending=False)
print(ft.head(10))
```

```python
y_test_pred = model.predict(val_x)
a_test_predict = np.where(y_test_pred > 0.5, 1, 0)
from sklearn.metrics import classification_report
print(classification_report(val_y, a_test_predict))
```

```python
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(val_y, a_test_predict)
```

```python
from sklearn.metrics import f1_score
f1_score(val_y,a_test_predict)
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(val_y, a_test_predict)
```

```python
def gini(p):
    return (p)*(1 - (p)) + (1 - p)*(1 - (1-p))

y_keys = list(val_y.keys())
y_val = []
for i in y_keys:
    y_val.append(val_y[i])

print("Gini Index =",gini(y_val.count(0)/len(y_val)))
```
